```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
X = iris.data
feature_names = iris.feature_names
```

```python
kmeans = KMeans(n_clusters=3, random_state=42) # 3 clusters because there are 3 classes of flowers in iris dataset
kmeans.fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
df = pd.DataFrame(X, columns=feature_names)
df['Cluster'] = labels
```

```python
print("=== K-Means Clustering Metrics ===")
print(f"Inertia: {kmeans.inertia_:.2f}")
print(f"Silhouette Score: {silhouette_score(X, labels):.3f}")
```

```python
print("\n=== Cluster Labels for Each Sample ===")
print(df[['Cluster']])
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, label='Samples')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.title("K-Means Clustering on Iris Dataset")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
